Самые популярные разновидности квантиля это медиана и квартили. Медиана (или 50-й квантиль/перцентиль) – это значение, меньше которого будет 50% всех значений. Квартили это квантили, которые делят значения на четверти (25%, 50% и 75%)


Метод, который объединяет в себе mean(), std(), min(), нахождение квартилей (в том числе второго квартиля – медианы) и max() – describe(). Идеальное решение, чтобы по-быстрому познакомиться со столбцом.

Группировка


Группировка данных – это объединение данных в группы по определенному признаку. Это полезный инструмент для нахождения закономерностей в разных группах.

Основная функция для этого в Pandas – groupby().
В круглые скобки мы заносим тот столбик, по которому будем группировать. Это всегда какой-то категориальный признак, по числовому признаку сгруппировать никак не получится!

df.groupby('Region')[['Sales', 'Profit']].sum() - группируем данные по регионам


df.groupby(['Region', 'Category'])[['Sales', 'Profit']].sum() - группируем по двум данным

В Pandas реализовано с помощью функции agg(). Она может принять в себя как одну функцию агрегирования, так и их список. Рассмотрим на примере

уменьшить размер таблицы, оставив только сводную статистику

Слияние данных – это процесс объединения двух (или более) таблиц на основе общих признаков. Это позволяет провести более комплексный анализ данных после совмещения информации из разных источников.

Один из способов слияния – merge(). Этот метод можно сравнить с SQL JOIN, так как они очень похоже работают. Вот основные его аргументы:

how – тип слияния (подробнее поговорим о них в следующем абзаце)
on – общий столбик, по которому будут присоединяться таблицы (использовать, если названия столбиков в двух таблицах одинаковые)
left_on – столбец из левого датасета, по которому следует проводить merge
right_on – столбец из правого датасета, по которому следует проводить merge
Аргументы left_on и right_on невозможно использовать вместе с on.

Перейдем к тому, какие существуют типы слияния(обозначим левый датасет за df1, а правый за df2):

left – в новом датасете сохраняются только те строки, которые есть в df1, даже если их нет в df2. Если данных нет, то значения будут заполнены NaN.
right – в новом датасете сохраняются только те строки, которые есть в df2, даже если их нет в df1. Если данных нет, то значения будут заполнены NaN.
inner – в новом датасете сохраняются только те строки, которые есть и в df1, и в df2. NaN из-за слияния появится не должно.
outer – в новом датасете сохранятся все строки, но если каких-то данных не будет, то появятся NaN.

```python
import pandas as pd
import numpy as np
dfp1 = pd.DataFrame({
    'Employee_ID': [1, 2, 3],
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Salary': [50000, 60000, 70000]
})

# Второй DataFrame с информацией о бонусах
dfp2 = pd.DataFrame({
    'Employee_ID': [1, 2, 4],
    'Name': ['Alice', 'Bob', 'David'],
    'Salary': [10000, 15000, 20000]  # Это бонусы, а не основная зарплата
})
print(dfp1)
print()
print(dfp2)
# Объединяем данные по Employee_ID с использованием суффиксов
dfp = pd.merge(dfp1, dfp2, on='Employee_ID', suffixes=('_Main', '_Bonus'))
dfp.set_index('Employee_ID', inplace=True)
dfp
```

В него передается кортеж из двух строк – первая строка будет постфиксом добавляться к конфликтующим столбцам из левого датасета, а вторая к столбцам правого. У функций read_...(), таких как, например, read_csv() и read_excel() есть специальные аргументы, которые отвечают как раз за парсинг дат:

parse_dates= – аргумент указывает столбцу, какие столбцы нужно преобразовать в формат datetime. Принимает столбец/список столбцов, номер столбца/список номеров, из которых нужно сделать столбцы со временем.
date_parser= – сюда можно передать собственную функцию для парсинга даты.
format= – ничем не отличается от одноименного аргумента в to_datetime(). Принимает строку strftime и помогает Pandas определиться с форматом нашей строки.

Дату с помощью .dt.date
Время / .dt.time
Год / .dt.year

Ресэмплинг (resampling) – это процесс изменения частоты временного ряда. Он позволяет агрегировать данные на основе большего интервала (downsampling) или интерполировать данные на более частый интервал (upsampling):

Downsampling – метод обработки временных рядов, при котором с помощью агрегации уменьшается частота временного ряда. Мы снижаем частоту: берём данные, например, раз в день, а делаем раз в месяц.
Upsampling – метод обработки временных рядов, при котором с помощью интерполяции увеличивается частота временного ряда. Мы увеличиваем частоту: были данные раз в месяц, хотим раз в день.
В pandas это делается через функцию resample(), в скобках вы указываете правило ресэмплинга (его частоту) с помощью строки. Вот какие основные значения он может принимать:

'S' – секунды.
'T' или 'min' - минуты.
'H' – часы.
'W' – недельные данные (неделя кончается в вс).
'M' – конец месяца.
'MS' – начало месяца.
'Q' – конец квартала
'QS' – начало квартала
'A' или 'Y' – конец года
'AS' 'YS' – начало года
'L' или 'ms' – миллисекунды
'U' – микросекунды
Важно: ресэмплинг производится с пмомощью индексов, поэтому перед ресэмплингом нужно поставить ваш столбец со временем как индекс



Интерполяция — это способ заполнить пропущенные значения,
предполагая, как могли бы вести себя данные между известными точками.
interpolate(method='linear') - Метод 'linear' просто соединяет точки прямой линией и ставит значение по центру.

Интерполяция ≠ угадывание, это просто разумное предположение

Метод должен соответствовать типу данных и контексту задачи

Метод (method=) | Что делает | Когда полезен
'linear' | Прямая между соседними точками | Самый универсальный, быстрый
'time' | Учитывает временные метки (если индекс — даты) | Когда интервалы неравномерные
'polynomial' | Аппроксимация полиномом (можно задать степень: order=2) | Плавная, более точная аппрокс.
'spline' | Похож на полином, но мягче (разбивает на куски) | Гладкое восстановление
'nearest' | Просто копирует ближайшее известное значение | Когда значения дискретны
'pad' (ffill) | Копирует предыдущее значение вниз | «Удерживает» последнее известное
'backfill' (bfill) | Копирует следующее значение вверх | «Заполняет будущее» назад
