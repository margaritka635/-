```python
def ini_sex(name):
    if "Mr." in name:
        return "male"
    elif "Mrs." in name or "Miss." in name: 
        return "female"
    else:
        return "male"
    return "unknown"
```

```python
data.loc[data["Sex"] == "unknown", "Sex"] = data["Name"].apply(ini_sex)
```
