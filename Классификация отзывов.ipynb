```python
import pandas as pd
import re
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, f1_score

df = pd.read_csv('amazon_review.csv') 
```

```python
df
```

```python
df = df[df['overall'] != 3.0]
df['label'] = df.loc[:, 'overall'].apply(lambda x: 1 if x > 3.0 else 0)
```

```python
def preprocess_text(text):
    if isinstance(text, str):
        text = text.lower()
        text = re.sub(r'[^a-zа-яё0-9\s]', '', text)
        text = re.sub(r'\s+', ' ', text).strip()
    else:
        text = ''
    return text
df['reviewText'] = df['reviewText'].apply(preprocess_text)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    df['reviewText'], df['label'], test_size=0.2, random_state=42
)
```

```python
vectorizer = TfidfVectorizer(max_features=10000)
X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)
```

```python
model = LogisticRegression(max_iter=1000)
model.fit(X_train_vec, y_train)
```

```python
y_pred = model.predict(X_test_vec)

accuracy = accuracy_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print(f'Accuracy: {accuracy:.4f}')
print(f'F1 Score: {f1:.4f}')
```

```python
модель отлично различает положительные и отрицательные отзывы и практически не путается.
```
